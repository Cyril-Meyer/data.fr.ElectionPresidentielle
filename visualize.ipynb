```python
import pandas as pd
import pygal

import pyperclip as pc

import matplotlib

from IPython.display import HTML
from IPython.display import SVG

import matplotlib
```

```python
cmap_gauche = matplotlib.cm.get_cmap('Reds')
cmap_droite = matplotlib.cm.get_cmap('Blues')
```

```python
undefined_color = []
for i in range(100):
    undefined_color.append("#ffffff")
undefined_color = tuple(undefined_color)

undefined_color_style = pygal.style.Style(colors=undefined_color)
```

```python
import math
def code_to_pygal(code):
    if code == 'ZA':
        return '971'
    elif code == 'ZB':
        return '972'
    elif code == 'ZC':
        return '973'
    elif code == 'ZD':
        return '974'
    elif code == '2A':
        return '2A'
    elif code == '2B':
        return '2B'
    elif type(code) == float:
        return None
    elif code[0] == 'Z':
        return None
    elif int(code) < 10:
        return f"0{code}"
    return code
```

```python
data_df_1995_t1 = pd.read_csv('1995/tour_1.csv')
```

```python
chart = pygal.maps.fr.Departments(show_legend=False, style=undefined_color_style)
chart.title = '1995 - CHIRAC'

for i in list(data_df_1995_t1['Code du département']):
    df = data_df_1995_t1.loc[data_df_1995_t1['Code du département'] == i]
    
    droite = 0
    gauche = 0

    # VILLIERS DE
    droite += df['Voix'].sum()
    # LE PEN
    droite += df['Voix.1'].sum()
    # CHIRAC
    droite += df['Voix.2'].sum()
    # CHEMINADE
    droite += df['Voix.4'].sum()
    # BALLADUR
    droite += df['Voix.7'].sum()

    # LAGUILLER
    gauche += df['Voix.3'].sum()
    # JOSPIN
    gauche += df['Voix.5'].sum()
    # VOYNET
    gauche += df['Voix.6'].sum()
    # HUE
    gauche += df['Voix.8'].sum()
    
    position = (droite - gauche ) / df['Exprimés'].sum()
    position = abs(position)
    position *= 1
    
    if gauche > droite:
        r, g, b, a = cmap_gauche(position)
    elif droite > gauche:
        r, g, b, a = cmap_droite(position)
    r *= 255
    g *= 255
    b *= 255
    
    chart.add(f"{code_to_pygal(i)}", [{'value':f"{code_to_pygal(i)}", 'color': f'rgb({r},{g},{b})'}])
    
svg = chart.render()
chart.render_to_file('1995_t1.svg')
SVG(svg)
```
